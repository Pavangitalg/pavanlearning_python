#**Learning matplotlib in the Python in this tutorial**

```python
import matplotlib.pyplot as plt
# Hours Vs Marks
# Hours = [2,4,6,8,10]
# marks = [21,44,76,23,97]


Hours = [2,4,6,8,10]
math_marks = [26,44,52,73,97]
sci_marks= [25,49,58,77,96]
plt.scatter(Hours,math_marks,label='Math')
plt.scatter(Hours,sci_marks,label='Science')
plt.xlabel("Hours Spent")
plt.ylabel("Marks Score")
plt.title("My first plot for Study Vs Marks")
plt.legend()
plt.show()
```
